```python
def rmse(predictions, targets):

    differences = predictions - targets                       #the DIFFERENCEs.

    differences_squared = differences ** 2                    #the SQUAREs of ^

    mean_of_differences_squared = differences_squared.mean()  #the MEAN of ^

    rmse_val = np.sqrt(mean_of_differences_squared)           #ROOT of ^

    return rmse_val   
```

```python
def log_trans(df, log_col_list):
    for column in log_col_list:
        df[column]=df[column].apply(lambda x: 0 if x==0 else np.log10(x))
    return df
```

```python
def year_transform(df, years_column):
    for column in years_column:
        df[column]=df[column]-(df[column].min()-1)
        df[column]=df[column].apply(lambda x: 0 if x==0 else np.log10(x))
    return df
```
